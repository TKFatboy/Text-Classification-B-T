```python
import pandas as pd
from pythainlp.tokenize import word_tokenize
from IPython.display import display
import os

# ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤ Path ‡πÑ‡∏ü‡∏•‡πå
file_name = r"d:\year4\‡∏™‡∏´‡∏Å‡∏¥‡∏à\prachatai_test.csv"

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÅ‡∏•‡∏∞‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå
if not os.path.exists(file_name):
    print(f"‚ùå Error: file not found '{file_name}'")
else:
    df = pd.read_csv(file_name)
    print(f"‚úÖ ‡πÇ‡∏´‡∏•‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à! ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {len(df)} ‡πÅ‡∏ñ‡∏ß")
    
    # ‡πÉ‡∏ô Jupyter ‡πÄ‡∏£‡∏≤‡πÉ‡∏ä‡πâ display(df.head()) ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏ï‡∏≤‡∏£‡∏≤‡∏á‡∏™‡∏ß‡∏¢‡∏Å‡∏ß‡πà‡∏≤ print
    display(df.head(3))
```

```python
# ===========================
# üîç ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏î‡∏π‡∏Ç‡πà‡∏≤‡∏ß‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ (‡πÅ‡∏Å‡πâ‡πÄ‡∏•‡∏Ç‡∏ï‡∏£‡∏á‡∏ô‡∏µ‡πâ!)
idx = 0  
# ===========================

if idx < len(df):
    news_content = df['body_text'][idx]
    news_title = df['title'][idx]

    print(f"--- üì∞ ‡∏´‡∏±‡∏ß‡∏Ç‡πâ‡∏≠‡∏Ç‡πà‡∏≤‡∏ß (Index {idx}) ---")
    print(f"{news_title}")
    print(f"\n--- üìù ‡πÄ‡∏ô‡∏∑‡πâ‡∏≠‡∏´‡∏≤‡∏Ç‡πà‡∏≤‡∏ß (200 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÅ‡∏£‡∏Å) ---")
    print(f"{news_content[:200]}...") # ‡∏ï‡∏±‡∏î‡∏°‡∏≤‡πÇ‡∏ä‡∏ß‡πå‡∏™‡∏±‡πâ‡∏ô‡πÜ ‡∏Å‡∏±‡∏ô‡∏£‡∏Å
    
    # --- ‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Å‡∏≤‡∏£‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥ ---
    tokens = word_tokenize(news_content, engine='newmm')
    
    print(f"\n" + "="*30)
    print(f"‚úÇÔ∏è ‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏Å‡∏≤‡∏£‡∏ï‡∏±‡∏î‡∏Ñ‡∏≥ (Total: {len(tokens)} ‡∏Ñ‡∏≥)")
    print("="*30)
    print(tokens[:30]) # ‡∏î‡∏π 30 ‡∏Ñ‡∏≥‡πÅ‡∏£‡∏Å
else:
    print("‚ùå ‡πÄ‡∏•‡∏Ç Index ‡πÄ‡∏Å‡∏¥‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ")
```

```python
# ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡∏´‡∏°‡∏ß‡∏î‡∏´‡∏°‡∏π‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
labels = ['politics', 'human_rights', 'quality_of_life', 'international', 
          'social', 'environment', 'economics', 'culture', 'labor', 
          'national_security', 'ict', 'education']

# ‡∏î‡∏∂‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏¢‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏î‡∏π
current_categories = []
for label in labels:
    if df[label][idx] == 1: # ‡πÉ‡∏ä‡πâ‡∏Ñ‡πà‡∏≤ idx ‡∏à‡∏≤‡∏Å Cell ‡∏ó‡∏µ‡πà‡πÅ‡∏•‡πâ‡∏ß
        current_categories.append(label)
        
print(f"‚úÖ ‡∏Ç‡πà‡∏≤‡∏ß‡∏ô‡∏µ‡πâ (Index {idx}) ‡∏ñ‡∏π‡∏Å‡∏à‡∏±‡∏î‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏´‡∏°‡∏ß‡∏î‡∏´‡∏°‡∏π‡πà:")
print(f"-> {current_categories}")
```
